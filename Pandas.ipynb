```python
# index 형태로 변환 가능한 표현식은 [] 안에 넣기 가능 -> 별로 안 좋음 헷갈리기만 함
print(df[0:10])
```

```python
# boolean indexing []안에 가능
print(df[df['Age']>20])
```

```python
# Dataframe.iloc[] -> 위치 기반 인덱싱
print(df.iloc[0:10,0:3])
```

```python
#Dataframe.loc[] -> 명칭 기반 인덱싱
print(df.loc[0:10,['Age','Pclass']])
```

```python
# boolean 인덱싱 -> iloc은 booelean indexing 불가, loc는 가능
print(df[df['Age']>60].head())
```

```python
print(df[df['Age']>60][['Survived','Pclass','Name']].head())
```

```python
print(df.loc[df['Age']>60,['Survived','Pclass','Name']].head())
```

```python
print(df[(df['Age']>60) & (df['Pclass']==1)].head())
```

```python
print(df[(df['Age']>60) | (df['Pclass']==1)].head())
```

# 결손 데이터 처리하기

```python
df.isna().head()
```

```python
df.isna().sum()
```

```python
df['Cabin']=df['Cabin'].fillna('C000') # inplace 가능
df.head()
```

```python
df['Age']=df['Age'].fillna(df.groupby(by='Sex')['Age'].mean())
df.head()
```

```python
lam = lambda x : x**2
print(lam(3))
```

```python
a=[1,2,3]
sqrs = map(lambda x: x**2,a)
print(list(sqrs))
```

```python
df['name_len']=df['Name'].apply(lambda x: len(x))
df.head()
```

```python
df['CorA'] = df['Age'].apply(lambda x: 'Child' if x<15 else 'Adult')
df.head()
```

```python
def get_cat(age):
    if age<15:
        return 'Child'
    elif age<30:
        return 'Adult'
    else:
        return "Sliver"
```

```python
df['CorA2']=df['Age'].apply(get_cat)
df.head()
```
